##### Primero veremos la correlación que existe entre los diferentes campos de datos: 

```python
f,ax = plt.subplots(1,1,figsize=(10,10))
ax = sns.heatmap(lista.corr(),annot=True)
plt.show()
```

Análisis: Como podemos ver el número de suicidios tiene una correlación de -0.0045 lo que significa que el número de suicidios no depende del año en el que sucedieron, podemos ver que la correlación entre el número de suicidios y la población es de 0.62, lo cual representa que a mayor número de población mayor es el número de suicidios, pero es un dato un poco irrelevante y un poco evidente al igual que el número de suicidios con los suicidios por cada 100k habitantes, así que la correlación más importante sería con 0.15 el número de suicidios depende en gran medida por el índice de desarrollo humano, y en menor medida por el producto interno bruto per cápita lo que nos hace pensar que el suicidio depende mucho de la economía y el desarrollo de un país.

##### Ahora analizaremos la contribucion de cada rango de edad en el conteo de suicidios por año.

```python
def count_year(group):
    return len(group.year.unique())
country_o = lista.groupby('country').apply(count_year).sort_values(ascending = False)
country_o.name = 'occurances of country'
nmaster = lista[~ (lista.year == 2016)].drop(['HDI for year','country-year','suicides/100k pop'],axis = 1)
coun_year = lista.groupby('country-year').population.sum()
ncountry_o = nmaster.groupby('country').apply(count_year).sort_values(ascending = False)
ncountry_o[country_o >= 30] 
pmaster = nmaster[nmaster.country.isin(ncountry_o[country_o >= 31].index)]
age_group = pd.pivot_table(pmaster,index=['year'],values='suicides_no',columns=['age'],aggfunc=np.sum)
sui_age=age_group[['5-14 years','15-24 years', '25-34 years', '35-54 years','55-74 years', '75+ years']]
sui_age.plot(kind = 'bar',figsize=(10,6),stacked=True,title='How each age groups contribute to the count of suicides')
```

Análisis: Como podemos ver la contribución de cada edad para el conteo es casi la misma en todos los años, pero a partir del año 2000 el aporte al conteo es más grande en el rango de 35 a 54 años de edad y se puede ver que tiende a crecer mientras avanzan los años. 

##### El numero de suicidios por edad y por genero 

```python
p = sns.barplot(x='sex', y='suicides_no', hue='age', data=lista)
```

Análisis: Como se puede observar la mayor cantidad de suicidios ocurre en el genero masculino, ademas de esto la mayor cantidad de encuentra en la edad de 35 a 54 años, ademas de esto podemos ver que las mujeres son menos suceptibles a los suicidios.

##### Numero de suicidios por cada 100 mil habitantes deacuerdo a su edad

```python
plt.figure(figsize=(10,5))
sns.barplot(x = "age", y = "suicides/100k pop", hue = "sex",data = lista.groupby(["age","sex"]).sum().reset_index()).set_title("Edad vs Suicidios")
plt.xticks(rotation = 90)
```

Análisis: Se puede apreciar que el mayor numero de suicidios por cada 100k habitantes es de personas de mas de 75 años, este dato es mas real que el anterior ya que es un indice que no toma en cuenta la cantidad sino el porcentaje de suicidios.

##### Paises con mas porcentaje de suicidios (Top 10)

```python
data_suicide_mean = lista['suicides/100k pop'].groupby(lista.country).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_suicide_mean.head(10).index,data_suicide_mean.head(10),palette='coolwarm')
```

Análisis: Podemos ver que el porcentaje mas alto de suicidios lo tiene Lituania, el cual es un pais cuya deuda por persona es de 6000 la cual puede ser una de las razones por las cuales esta al inicio de esta lista.

##### Numero de suicidios por año y su realacion con el producto interno bruto per capita.

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(lista.groupby(['year']).suicides_no.sum(),'ko--')
tick = ax.set_xticks(range(1985,2018,2))
ax.set_title('Total suicidios de 1985 a 2016')
ax.set_xlabel('Año')
ax.set_ylabel('Suicidios')

#Grafico 1
```

```python
data_gdp = (lista['gdp_per_capita ($)'].groupby(lista.year)).sum()
data_gdp.plot(figsize=(20,10), linewidth=2, fontsize=15,color='red')
plt.xlabel('Year', fontsize=15)
plt.ylabel(' Total gdp_per_capita ($)',fontsize=15)
plt.show()


#Grafico 2
```

Análisis: Como podemos ver en el grafico 1 se puede observar una tendencia a subir el numero de suicidios mientras avanza el tiempo, pero esto no es solo por el cambio de epoca sino tambien pueden intervenir varios factores como el crecimiento de la poblacion, y factores sociales, en este caso analizaremos la relacion con el producto interno bruto per capita por año, asi podemos ver que si bien las graficas no son identicas sufren la misma tendencia, se puede concluir que el aspecto economico es uno de los factores fundamentales para el crecimiento o no del incide de suicidios a nivel mundial.

##### Producto interno bruto per capita vs suicidios por 100k habitantes

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.regplot(x='gdp_per_capita ($)', y='suicides/100k pop',data=lista)
plt.show()
```

Análisis: Como lo vimos en el análisis anterior queriamos asegurarnos que esa conclucion fue correcta, y lo corroboramos al ver como es el comportamiento de este indice minimizando los factores ajenos al producto interno bruto, como pudo ser el numero de habitantes, ese factor queda anulado al manejarnos con porcentaje de numero de suicidios por cada 100k habitantes y obtenemos que a menor PIB por habitante la tasa de suicidios es mas grande asi que podemos asegurar que el factor economico es determinante para el numero de suicidios
