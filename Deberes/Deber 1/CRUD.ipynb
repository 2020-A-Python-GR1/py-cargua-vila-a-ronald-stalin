```python
import copy

# Clase Archivo

class File:
    
    # Create file
    def create(self,path,value):  
        try:
            open_file = open(path, 'a')
            open_file.write(value)
            open_file.close()
        except Exception as error:
            print("Error at reading file - Create")

    # Read file
    def read(self,path):  
        try:
            open_file = open(path, 'r')
            file = open_file.read()               
            open_file.close()
            return file
        except Exception as error:
            print("Error at reading file - Read")
            
    # Read Update file
    def read_para_update(self,path):  
        try:
            open_file = open(path, 'r')
            file = open_file.readlines()               
            open_file.close()
            return file
        except Exception as error:
            print("Error at reading file - Read Update")
            
            
    # Update file
    def update(self,path, nuevo_valor, indice_de_la_fila):  
        
        try:
            open_file = open(path, "r")
            list_of_lines = open_file.readlines()
            list_of_lines[indice_de_la_fila] = nuevo_valor

            open_file = open(path, "w")
            open_file.writelines(list_of_lines)
            open_file.close()
        except Exception as error:
            print("Error at reading file - Update")
            
            
    # Delete File
    def delete(self, path, a_eliminar):  
       
        try:
            open_file = open(path, "r")
            list_of_lines = open_file.readlines()
            lines_to_write = []
            for line in list_of_lines:
                if line != a_eliminar:
                    lines_to_write.append(line)

            open_file = open(path, "w")
            open_file.writelines(lines_to_write)
            open_file.close()
        except Exception as error:
            print("Error at reading file - Delete")    

    
nuevo_file = File()




# Clase Equipo
path_equipo = './equipos.txt'

class Equipo:
    
    # Todos los atributos públicos
    nombre = None
    titulos = None
    anio = None
    champions = None
    entrenador = None
    jugadores = None
    
    # Create
    def __init__(self, nombre, titulos, anio, champions, entrenador):
        self.nombre = nombre
        self.titulos = titulos
        self.anio = anio
        self.champions = champions
        self.entrenador = entrenador
        self.jugadores = []
        equipo = self.__str__()
        nuevo_file.create(path_equipo,equipo)
                
        
    # Read
    def read_equipo(self):
        data = nuevo_file.read(path_equipo)
        print(data)
        return data
                
        
    # Añadir y eliminar una jugador al registro de un País.
    def jugadores_add_delete(self, nombre_equipo, jugador_involucrada, add_o_no=True):
        add = str(add_o_no)
        lista_equipos = nuevo_file.read_para_update(path_equipo)
        for equipo in lista_equipos:
            data = equipo.split(";")
            if nombre_equipo in data:
                indice_fila = lista_equipos.index(equipo)
                equipo_modificado = copy.copy(data)
                aux = copy.copy(equipo_modificado[5])
                
                data_jugador = aux.strip("']['").split(', ') # retorna lista de las jugadores en ese equipo

                if add == "True":
                    lista_jugadores = nuevo_file.read_para_update(path_jugador)
                    encontrado = False
                    for jugador in lista_jugadores:
                        data = jugador.split(";")
                        if jugador_involucrada in data:
                            encontrado = True
                            data_jugador.append(jugador_involucrada)    # lista                               
                            equipo_modificado[5] = copy.copy(data_jugador)
                            aux_jugadores = "_".join(equipo_modificado[5]) # str
                            equipo_modificado[5] = copy.copy(aux_jugadores) 
                            modificado = ";".join(equipo_modificado)
                            nuevo_file.update(path_equipo, modificado, indice_fila)
                            break

                    if encontrado == False:
                        msg = '{} no existe en el archivo jugadores'.format(jugador_involucrada)
                        print(msg)
                        return msg
                            
                else: # delete a city in a coutnry
                    lista_jugadores = nuevo_file.read_para_update(path_jugador)
                    encontrado = False
                    for jugador in lista_jugadores:
                        data = jugador.split(";")
                        if jugador_involucrada in data:
                            encontrado = True
                            dataC = "_".join(data_jugador)   
                            dataC1  = dataC.split("_")
                            dataC1.remove(jugador_involucrada)
                            equipo_modificado[5] = copy.copy(dataC1)
                            aux_jugadores = "_".join(equipo_modificado[5])
                            equipo_modificado[5] = aux_jugadores 
                            modificado = ";".join(equipo_modificado)
                            nuevo_file.update(path_equipo, modificado, indice_fila)
                            break

                    if encontrado == False:
                        msg = '{} no existe en el archivo jugadores'.format(jugador_involucrada)
                        print(msg)
                        return msg                   
                
                
                nuevo_equipo.read_equipo()

        
    # Update
    def update_equipo(self, nombre_equipo,atributo, nuevo_valor):
    
        equipo_modificado = None
        indice_fila = None
        lista_equipos = nuevo_file.read_para_update(path_equipo)
        
        for equipo in lista_equipos:
            data = equipo.split(";")
            if nombre_equipo in data:
                indice_fila = lista_equipos.index(equipo)
                equipo_modificado = copy.copy(data)
                break
        
        if equipo_modificado == None:
            msg = ' {} no encontrado.'.format(nombre_equipo)
            print(msg)
            return msg
                  
        def u_nombre():
            equipo_modificado[0] = nuevo_valor
            return actualizar()
            
        def u_titulos():
            equipo_modificado[1] = nuevo_valor
            return actualizar()
        
        def u_anio():
            equipo_modificado[2] = nuevo_valor
            return actualizar()
            
        def u_champions():
            equipo_modificado[3] = nuevo_valor
            return actualizar()
            
        def u_entrenador():
            equipo_modificado[4] = nuevo_valor
            return actualizar()          
        
            
        def actualizar():
            modificado = ";".join(equipo_modificado)
            nuevo_file.update(path_equipo, modificado, indice_fila)
            nuevo_equipo.read_equipo()
            
            
        def seleccionar():
            opciones = {
                "nombre" : u_nombre,
                "titulos": u_titulos,
                "anio": u_anio,
                "champions": u_champions,
                "entrenador": u_entrenador,
            }
            
            return opciones[atributo]()
            
        return seleccionar()
    


    # Delete equipo
    def delete_equipo(self, nombre_equipo_a_eliminar):

        equipo_a_eliminar = None
        lista_equipos = nuevo_file.read_para_update(path_equipo) # tambien sirve para eliminar
        
        for equipo in lista_equipos:
            data = equipo.split(";")
            if nombre_equipo_a_eliminar in data:
                equipo = copy.copy(data)
                equipo_a_eliminar = ";".join(equipo)
                break
    
        nuevo_file.delete(path_equipo, equipo_a_eliminar)
        
    
    # To string
    def __str__(self):
        string = '{};{};{};{};{};{}; \n'.format(self.nombre,self.titulos,self.anio,self.champions,
                                               self.entrenador,self.jugadores)
        return string

    
    
    
# Clase Jugador

path_jugador = './jugadores.txt'

class Jugador:
    
    # Todos los atributos públicos
    nombre = None
    apellido = None
    edad = None
    numero = None
    posicion = None
    
    # Create jugador
    def __init__(self, nombre, apellido, edad, numero, posicion):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.numero = numero
        self.posicion = posicion
        jugador = self.__str__()
        nuevo_file.create(path_jugador,jugador)
        
        
        
    # Read
    def read_jugador(self):
        data = nuevo_file.read(path_jugador)
        print(data)
        return data
        
        
        
    # Update jugador
    def update_jugador(self, nombre_jugador,atributo, nuevo_valor):
    
        jugador_modificada = None
        indice_fila = None
        lista_jugadores = nuevo_file.read_para_update(path_jugador)
        
        for jugador in lista_jugadores:
            data = jugador.split(";")
            if nombre_jugador in data:
                indice_fila = lista_jugadores.index(jugador)
                jugador_modificada = copy.copy(data)
                break

                  
        def u_nombre():
            jugador_modificada[0] = nuevo_valor
            return actualizar()
            
        def u_apellido():
            jugador_modificada[1] = nuevo_valor
            return actualizar()
        
        def u_edad():
            jugador_modificada[2] = nuevo_valor
            return actualizar()
            
        def u_numero():
            jugador_modificada[3] = nuevo_valor
            return actualizar()
            
        def u_posicion():
            jugador_modificada[4] = nuevo_valor
            return actualizar()
            
            
        def actualizar():
            modificada = ";".join(jugador_modificada)
            nuevo_file.update(path_jugador, modificada, indice_fila)
            nuevo_jugador.read_jugador()
            
            
        def seleccionar():
            opciones = {
                "nombre" : u_nombre,
                "apellido": u_apellido,
                "edad": u_edad,
                "numero": u_numero,
                "posicion": u_posicion
            }
            
            return opciones[atributo]()
            
        return seleccionar()
    

    
    # Delete jugador
    def delete_jugador(self, nombre_jugador_a_eliminar):

        jugador_a_eliminar = None
        lista_jugadores = nuevo_file.read_para_update(path_jugador) # tambien sirve para eliminar
        
        for jugador in lista_jugadores:
            data = jugador.split(";")
            if nombre_jugador_a_eliminar in data:
                jugador = copy.copy(data)
                jugador_a_eliminar = ";".join(jugador)
                break
    
        nuevo_file.delete(path_jugador, jugador_a_eliminar)
        
    
    
    # To string
    def __str__(self):
        string = '{};{};{};{};{}; \n'.format(self.nombre,self.apellido,self.edad,
                                                self.numero, self.posicion)
        return string

    

def operaciones_equipo(atributo):
    
    def crear_equipo():
        nombre = input('\n Ingrese el nombre del equipo:')
        titulos = input('Ingrese los titulos:')
        anio = input('Ingrese el año:')
        champions = input('Ingrese la champions (True o False):')
        entrenador = input('Ingrese el entrenador')
        nuevo_equipo = Equipo(nombre, titulos, anio, champions, entrenador)
                
    def ver_equipos():
        nuevo_equipo.read_equipo()
        
    def actualizar_atributos_equipo():
        nombre_equipo = input('\n Ingrese el nombre del equipo a actualizar:')
        atributo = input('Ingrese el atributo a actualizar:')
        nuevo_valor = input('Ingrese el nuevo valor:')
        nuevo_equipo.update_equipo(nombre_equipo, atributo, nuevo_valor)
        
    def anadir_delete():
        nombre_equipo = input('\n Ingrese el nombre del equipo:')
        jugador_involucrada = input('Ingrese el nombre de la jugador:')
        add_delete = input('Ingrese "True" para añadir o "False" para eliminar:')
        nuevo_equipo.jugadores_add_delete(nombre_equipo, jugador_involucrada, add_delete)
        
    def eliminar_equipo():
        nombre_equipo = input('\n Ingrese el nombre del equipo a eliminar:')
        nuevo_equipo.delete_equipo(nombre_equipo)
        
    def decision():
        
        opciones = {
            "1": crear_equipo,
            "2": ver_equipos,
            "3": actualizar_atributos_equipo,
            "4": anadir_delete,
            "5": eliminar_equipo
        }
        
        return opciones[atributo]()
        
    return decision()
    
    
def operaciones_jugador(atributo):

    def crear_jugador():
        nombre = input('\n Ingrese el nombre del jugador:')
        zone = input('Ingrese la apellido del jugador:')
        edad = input('Ingrese el edad del jugador:')
        numero = input('Ingrese la numero :')
        posicion = input('Ingrese la posicion:')
        nuevo_jugador = Jugador(nombre, zone, edad, numero, posicion)
                
    def ver_jugadores():
        nuevo_jugador.read_jugador()
        
    def actualizar_atributos_jugador():
        nombre_jugador = input('\n Ingrese el nombre del jugador a actualizar:')
        atributo = input('Ingrese el atributo a actualizar:')
        nuevo_valor = input('Ingrese el nuevo valor:')
        nuevo_jugador.update_jugador(nombre_jugador, atributo, nuevo_valor)
        
    def eliminar_jugador():
        nombre_jugador = input('\n Ingrese el nombre del jugador a eliminar:')
        nuevo_jugador.delete_jugador(nombre_jugador)
        
    def decision():
        
        opciones = {
            "1": crear_jugador,
            "2": ver_jugadores,
            "3": actualizar_atributos_jugador,
            "4": eliminar_jugador
        }
        
        return opciones[atributo]()
        
    return decision()
    
def iniciar():
    clase = input('Equipo(1) o Jugador(2): ')
    decision = 0
    
    if clase == "1": # Equipo
        decision = input(' \n 1) Crear Equipo \n 2) Ver Equipos \n 3) Actualizar atributos de un equipo \n 4) Añadir o eliminar una jugador a un equipo \n 5) Eliminar equipo: ')
        operaciones_equipo(decision)
        
        
    else: # Jugador
        decision = input(' \n 1) Crear Jugador \n 2) Ver Jugadores \n 3) Actualizar atributos de una jugador \n 4) Eliminar jugador: ')
        operaciones_jugador(decision)
```
